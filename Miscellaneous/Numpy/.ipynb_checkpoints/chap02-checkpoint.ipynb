```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.misc
%pylab inline
```

### Resizing Images

```python
face = scipy.misc.face()

# Check the shape of the face array
face.shape
```

```python
# Resize
resized = face.repeat(6, axis=0).repeat(5, axis=1)
```

```python
# Plot the face image 
plt.subplot(211)
plt.title("Face")
plt.axis("off")
plt.imshow(face)

# Plot the resized face image
plt.subplot(212)
plt.title("Face Resized")
plt.axis("off")
plt.imshow(resized)
plt.show()
```

### Creating Views and Copies

```python
acopy = face.copy()
aview = face.view()

# Plot the face array
plt.subplot(221)
plt.imshow(face)
plt.title("Face array")

# Plot the copy
plt.subplot(222)
plt.imshow(acopy)
plt.title("The copy")


# Plot the view
plt.subplot(223)
plt.imshow(aview)
plt.title("The view")

# Plot the view after changes
aview.flat = 0
plt.subplot(224)
plt.imshow(aview)
plt.title("The View after changes")
plt.show()
```

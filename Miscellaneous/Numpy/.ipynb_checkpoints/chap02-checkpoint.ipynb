```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.misc
%pylab inline
```

### Resizing Images

```python
ascent = scipy.misc.ascent()

# Check the shape of the ascent array
ascent.shape
```

```python
# Resize
resized = ascent.repeat(6, axis=0).repeat(5, axis=1)
```

```python
# Plot the ascent image 
plt.subplot(211)
plt.title("ascent")
plt.axis("off")
plt.imshow(ascent)

# Plot the resized ascent image
plt.subplot(212)
plt.title("ascent Resized")
plt.axis("off")
plt.imshow(resized)
plt.show()
```

### Creating Views and Copies

```python
acopy = ascent.copy()
aview = ascent.view()

# Plot the ascent array
plt.subplot(221)
plt.imshow(ascent)
plt.axis("off")
plt.title("ascent array")

# Plot the copy
plt.subplot(222)
plt.imshow(acopy)
plt.axis("off")
plt.title("The copy")


# Plot the view
plt.subplot(223)
plt.imshow(aview)
plt.axis("off")
plt.title("The view")

# Plot the view after changes
aview.flat = 0
plt.subplot(224)
plt.imshow(aview)
plt.axis("off")
plt.title("The View after changes")
plt.show()
```

### Flipping Images

```python
ascent = scipy.misc.ascent()

plt.imshow(ascent[:, ::-1])
# select top left quadrant of the picture

plt.imshow(ascent[:ascent.shape[0]/2, :ascent.shape[1]/2])
```

```python
# Plot the ascent array
plt.subplot(221)
plt.title('Original')
plt.axis('off')
plt.imshow(ascent)


# Plot the flipped array
plt.subplot(222)
plt.title('Flipped')
plt.axis('off')
plt.imshow(ascent[:,::-1])

# Plot a slice array
plt.subplot(223)
plt.title('sliced')
plt.axis('off')
plt.imshow(ascent[:ascent.shape[0]/2, :ascent.shape[1]/2])

# Apply a mask
mask = ascent % 2 == 0
masked_ascent = ascent.copy()
masked_ascent[mask] = 0
plt.subplot(224)
plt.title('Masked')
plt.axis('off')
plt.imshow(masked_ascent)

plt.show()
```

### Fancy indexing by setting values on the diagonals to 0

```python
ascent = scipy.misc.ascent()
xmax = ascent.shape[0]
ymax = ascent.shape[1]
```

```python
# Fancy indexing on diagonal to 0
# x 0-xmax
# y 0-ymax

ascent[range(xmax), range(ymax)] = 0
```

```python
# Set values on other diagonal to 0
# x xmax-0
# y 0-ymax

ascent[range(xmax-1,-1,-1), range(ymax)] = 0

# Plot ascent with diagonal lines set to 0
plt.imshow(ascent)
plt.show()
```

### Indexing with a list of locations 

```python
# Load the ascent array
ascent = scipy.misc.ascent()
xmax = ascent.shape[0]
ymax = ascent.shape[1]

def shuffle_indices(size):
    """Shuffles an array with values 0 - size """
    arr = np.arange(size)
    np.random.shuffle(arr)
    
    return arr

xindices = shuffle_indices(xmax)
np.testing.assert_equal(len(xindices), xmax)

yindices = shuffle_indices(ymax)
np.testing.assert_equal(len(yindices), ymax)

```

```python
# Plot ascent
plt.imshow(ascent[np.ix_(xindices, yindices)])
plt.show()
```

### Indexing with Booleans

```python
ascent = scipy.misc.ascent()

def get_indices(size):
    arr = np.arange(size)
    return arr % 4 == 0

# Plot Ascent
ascent1 = ascent.copy()
xindices = get_indices(ascent.shape[0])
yindices = get_indices(ascent.shape[1])
ascent1[xindices, yindices] = 0
plt.subplot(211)
plt.imshow(ascent1)

ascent2 = ascent.copy()
# Between quarter and 3 quarters of the max value

ascent2[(ascent > ascent.max() / 4) & (ascent < 3 * ascent.max() / 4)] = 0
plt.subplot(212)
plt.imshow(ascent2)

plt.show()
```

### Stride tricks for Sudoku

```python
sudoku = np.array([
[2, 8, 7, 1, 6, 5, 9, 4, 3],
[9, 5, 4, 7, 3, 2, 1, 6, 8],
[6, 1, 3, 8, 4, 9, 7, 5, 2],
[8, 7, 9, 6, 5, 1, 2, 3, 4],
[4, 2, 1, 3, 9, 8, 6, 7, 5],
[3, 6, 5, 4, 2, 7, 8, 9, 1],
[1, 9, 8, 5, 7, 3, 4, 2, 6],
[5, 4, 2, 9, 1, 6, 3, 8, 7],
[7, 3, 6, 2, 8, 4, 5, 1, 9]
])

shape = (3, 3, 3, 3)

strides = sudoku.itemsize * np.array([27, 3, 9, 1])

squares = np.lib.stride_tricks.as_strided(sudoku, shape = shape, strides = strides)

print squares
```

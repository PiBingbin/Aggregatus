```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.misc
%pylab inline
```

### Resizing Images

```python
face = scipy.misc.face()

# Check the shape of the face array
face.shape
```

```python
# Resize
resized = face.repeat(6, axis=0).repeat(5, axis=1)
```

```python
# Plot the face image 
plt.subplot(211)
plt.title("Face")
plt.axis("off")
plt.imshow(face)

# Plot the resized face image
plt.subplot(212)
plt.title("Face Resized")
plt.axis("off")
plt.imshow(resized)
plt.show()
```

### Creating Views and Copies

```python
acopy = face.copy()
aview = face.view()

# Plot the face array
plt.subplot(221)
plt.imshow(face)
plt.axis("off")
plt.title("Face array")

# Plot the copy
plt.subplot(222)
plt.imshow(acopy)
plt.axis("off")
plt.title("The copy")


# Plot the view
plt.subplot(223)
plt.imshow(aview)
plt.axis("off")
plt.title("The view")

# Plot the view after changes
aview.flat = 0
plt.subplot(224)
plt.imshow(aview)
plt.axis("off")
plt.title("The View after changes")
plt.show()
```

### Flipping Images

```python
face = scipy.misc.face()

plt.imshow(face[:, ::-1])
# select top left quadrant of the picture

plt.imshow(face[:face.shape[0]/2, :face.shape[1]/2])
```

```python
# Plot the face array
plt.subplot(221)
plt.title('Original')
plt.axis('off')
plt.imshow(face)


# Plot the flipped array
plt.subplot(222)
plt.title('Flipped')
plt.axis('off')
plt.imshow(face[:,::-1])

# Plot a slice array
plt.subplot(223)
plt.title('sliced')
plt.axis('off')
plt.imshow(face[:face.shape[0]/2, :face.shape[1]/2])

# Apply a mask
mask = face % 2 == 0
masked_face = face.copy()
masked_face[mask] = 0
plt.subplot(224)
plt.title('Masked')
plt.axis('off')
plt.imshow(masked_face)

plt.show()
```

### Fancy indexing by setting values on the diagonals to 0

```python
ascent = scipy.misc.ascent()
xmax = ascent.shape[0]
ymax = ascent.shape[1]
```

```python
# Fancy indexing on diagonal to 0
# x 0-xmax
# y 0-ymax

ascent[range(xmax), range(ymax)] = 0
```

```python
# Set values on other diagonal to 0
# x xmax-0
# y 0-ymax

ascent[range(xmax-1,-1,-1), range(ymax)] = 0

# Plot ascent with diagonal lines set to 0
plt.imshow(ascent)
plt.show()
```

### Indexing with a list of locations 

```python
# Load the ascent array
ascent = scipy.misc.ascent()
xmax = ascent.shape[0]
ymax = ascent.shape[1]

def shuffle_indices(size):
    """Shuffles an array with values 0 - size """
    arr = np.arange(size)
    np.random.shuffle(arr)
    
    return arr

xindices = shuffle_indices(xmax)
np.testing.assert_equal(len(xindices), xmax)

yindices = shuffle_indices(ymax)
np.testing.assert_equal(len(yindices), ymax)

```

```python
# Plot ascent
plt.imshow(ascent[np.ix_(xindices, yindices)])
plt.show()
```

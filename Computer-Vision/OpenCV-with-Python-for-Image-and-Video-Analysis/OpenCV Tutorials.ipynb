out = cv2.VideoWriter('output.avi', -1, 20.0, (640, 480))

    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Write the frame
    out.write(frame)

    cv2.imshow('frame', frame)
    cv2.imshow('gray', gray)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break


# Release everything if job is finished
cap.release()
out.release()

## 4. Image Arithmetic

```python

img1 = cv2.imread('hand.png')
img2 = cv2.imread('loc.png')
img3 = cv2.imread('bright.png')
#
# add = img2 + img3
# weighted = cv2.addWeighted(img2, 0.6, img3, 0.4, 0)
# cv2.imshow('add', add)
# cv2.imshow('weighted', weighted)

rows, cols, channels = img2.shape
roi = img1[0:rows, 0:cols]

# Create a mask of img 2
img2gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

# Apply the threshold
ret, mask = cv2.threshold(img2gray, 220, 255, cv2.THRESH_BINARY_INV)

mask_inv = cv2.bitwise_not(mask)

img1_bg = cv2.bitwise_and(roi, roi, mask=mask_inv)
img2_fg = cv2.bitwise_and(img2, img2, mask=mask)

dst = cv2.add(img1_bg, img2_fg)
img1[0:rows, 0:cols] = dst

plt.title('img2_fg')
plt.imshow(img2_fg)
plt.show()
```

## 5. Drawing

```python

img = cv2.imread('kigali.png', cv2.IMREAD_COLOR)

# Draw line on the image
cv2.line(img, (0, 0), (150, 150), (255, 150, 100), 15)


# Draw a rectangle
cv2.rectangle(img, (15, 25), (200, 150), (0, 255, 0), 5)


# Draw a circle
cv2.circle(img, (100, 78), 55, (0, 0, 255), -1)


# Draw a polygon
pts = np.array([[15, 5], [20, 30], [70, 20], [50, 10]], np.int32)
pts = pts.reshape((-1, 1, 2))
cv2.polylines(img, [pts], True, (0, 200, 200), 5)


# Write on the image
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img, 'KCC!', (0, 130), font, 1, (210, 210,15), 5)

plt.imshow(img)
plt.axis('off')
plt.show()
```

## 6. Thresholding

```python
img = cv2.imread('bookpage.png')
# set the threshold so that any pixel value less than 12 is set to black and any pixel valuel greater than 12 is set to white
retval, threshold = cv2.threshold(img, 12, 255, cv2.THRESH_BINARY)

plt.imshow(img)
plt.axis("off")
plt.title("Original")
plt.show()


plt.imshow(threshold)
plt.axis("off")
plt.title("with threshold")
plt.show()
```

```python
# Black and white threshold
img = cv2.imread('bookpage.png')
grayscaled = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
retval2, threshold2 = cv2.threshold(grayscaled, 12, 255, cv2.THRESH_BINARY)

plt.imshow(threshold2)
plt.axis("off")
plt.title("with Black and White threshold")
plt.show()
```
